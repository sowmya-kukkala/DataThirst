```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = pd.read_csv('C:/Users/SOWMYA/Desktop/Desktopfolder/DataScience Projects/Irisdata.txt',header = None)
```

```python
iris.head()
```

```python
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
iris.plot(x='sepal_length',y='sepal_width')
plt.show()
# Line Plot doesn't make that much sense in this scenario while you see in the below plot
```

```python
#Using Scatter Plot for the given variables/features of the Iris Data
iris.plot(x='sepal_length',y='sepal_width', kind = 'scatter')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
#Using the Box Plot to plot the various features of iris dataset
iris.plot(y = 'sepal_length', kind = 'box')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
#Using Histogram to plot the variable 'sepal length' of iris dataset
iris.plot(y='sepal_length', kind = 'hist')
plt.xlabel('sepal length (cm)')
plt.show()
```

```python
#customizing Histogram
iris.plot(y='sepal_length', kind='hist',bins= 30, range=(4,8),normed =True )
plt.xlabel('sepal length (cm)')
plt.show()
```

```python
# Cummulative Distribution
iris.plot(y='sepal_length', kind='hist', bins=30, range=(4,8),normed=True,cumulative=True,)
plt.xlabel('sepal length (cm)')
plt.title('Cummulative Distribution Function (CDF)')
plt.show()
```

```python
#Filtering by Species
indices = iris['species'] == 'Iris-setosa'

setosa = iris.loc[indices,:] # Extract new dataframe
```

```python
indices = iris.species == 'Iris-virginica'
virginica = iris.loc[indices,:] # Extract new dataframe
```

```python
#Visualize EDA (Exploratory Data Analysis) : all Data
iris.plot(kind='hist', bins = 50, range = (0,8), alpha = 0.3)
plt.title('Entire Iris Dataset')
plt.xlabel('[cm]')
plt.show()
```

```python
# Visualize EDA: Individual Factors
setosa.plot(kind = 'hist',bins = 50, range = (0,8), alpha = 0.3)
plt.title('Setosa data set')
plt.xlabel('[cm]')

versicolor.plot(kind = 'hist', bins = 50, range = (0,8), alpha = 0.3)
plt.title('Versicolor data set')
plt.xlabel('[cm]')

virginica.plot(kind = 'hist', bins = 50, range = (0,8), alpha = 0.3)
plt.title('virginica data set')
plt.xlabel('[cm]')

plt.show()
```

```python
#Statistical EDA : describe()
describe_all = iris.describe()
print(describe_all)
```

```python
describe_setosa = setosa.describe()
print(describe_setosa)
```

```python
describe_versicolor = versicolor.describe()
print(describe_versicolor)
```

```python
describe_virginica = virginica.describe()
print(describe_virginica)
```

```python
#Computing Errors
error_setosa = 100 * np.abs(describe_setosa - describe_all)
error_setosa = error_setosa/describe_setosa
print(error_setosa)
```
